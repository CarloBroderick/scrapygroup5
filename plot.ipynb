```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#import data
scraped_data = pd.read_csv('population.csv', na_values = "N.A.")

#change strings to numbers
scraped_data['fert_rate'] = scraped_data['fert_rate'].astype(float)
scraped_data['med_age'] = scraped_data['med_age'].astype(float)
```

```python
#apply default theme
sns.set_theme()

tutorial_plot = sns.relplot(
    data = scraped_data,
    x = 'fert_rate',
    y = 'med_age',
).set(title = 'Median Age by Fertility Rate')

tutorial_plot.set_ylabels(r'Median Age')
tutorial_plot.set_xlabels(r'Fertility Rate')

tutorial_plot
```
